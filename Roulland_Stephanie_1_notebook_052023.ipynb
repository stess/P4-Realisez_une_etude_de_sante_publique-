```python
#Modification de la colonne Année de sous_nutrition pour ne garder qu'une valeur
sous_nutrition["Year"] = sous_nutrition["Année"].apply(lambda x: int(x.split("-")[0]) + 1)
sous_nutrition.drop("Année", axis=1, inplace=True)
sous_nutrition.rename(columns={"Year": "Année"}, inplace=True)

# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017
personnes_sous_nutrition_2017 = pd.merge(population.loc[population['Année'] == 2017,["Zone", "Population"]],
                                sous_nutrition.loc[sous_nutrition['Année'] == 2017,["Zone", "sous_nutrition"]],
                                on='Zone')
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personne qui pourrait être nourries</h3>
</div>

```python
#Combien mange en moyenne un être humain ?
print("Besoins journaliers pour un homme : 2600 kcal/j")
print("Besoins journaliers pour une femme : 2100 kcal/j", end="\n\n")

moyenne_kcal = (2600 + 2100)/2

print("Besoins journaliers moyens pour un être humain : ", moyenne_kcal, " kcal/j")

#Source 1 : https://www.anses.fr/fr/system/files/NUT2012SA0103Ra-2.pdf (page 14)
#Source 2 : https://www.cerin.org/articles/references-nutritionnelles-pour-les-adultes/
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
dispo_alimentaire = dispo_alimentaire.merge(population.loc[population['Année'] == 2017,["Zone", "Population"]],
                                            on='Zone')
```

```python
#Affichage du nouveau dataframe
dispo_alimentaire.head()
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
dispo_alimentaire['dispo_kcal'] = dispo_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)'] * dispo_alimentaire['Population'] * 365

print("kcal disponibles mondialement :", dispo_alimentaire['dispo_kcal'].sum(), "kcal", end="\n\n")

#print(population.loc[population['Année'] == 2017,"Population"].sum())
```

```python
#Calcul du nombre d'humains pouvant être nourris
total_humains_kcal = round(dispo_alimentaire['dispo_kcal'].sum()/(moyenne_kcal*365))
print("Nombre humains pouvant être nourris en 2017 :", total_humains_kcal)

prop_humains_kcal = total_humains_kcal*100/population.loc[population['Année'] == 2017,"Population"].sum()
print("Proportion d'humains pouvant être nourris en 2017 :", "{:.2f}".format(prop_humains_kcal), "%")
```

```python
#création d'un dataframe avec uniquement la Thaïlande 
thailande_df = personnes_sous_nutrition_2017.loc[personnes_sous_nutrition_2017['Zone'] == 'Thaïlande',:]

display(thailande_df)
```

```python
#Calcul de la sous nutrition en Thaïlande
print('Proportion de la sous nutrition en Thaïlande :', "{:.2f}".format(thailande_df['sous_nutrition'].iloc[0]*100/thailande_df['Population'].iloc[0]), "%")
```

```python
proportion_manioc_exporte = dispo_alimentaire.loc[(dispo_alimentaire['Produit'] == "Manioc") & (dispo_alimentaire['Zone'] == "Thaïlande"),:]

# On affiche la quantité de manioc exporté de Thaïlande
quantite_manioc_exporte_thailande = proportion_manioc_exporte['Exportations - Quantité'].iloc[0]
print('Quantité de manioc exporté en Thaïlande : ', "{:.3f}".format(quantite_manioc_exporte_thailande/1000000), "millions de kilos", end="\n\n")

# On calcule la proportion exportée en fonction de la production
pourcentage_manioc_exporte = proportion_manioc_exporte['Exportations - Quantité'].iloc[0]*100 / proportion_manioc_exporte['Production'].iloc[0]
print('Proportion exportée de manioc en Thaïlande :', "{:.2f}".format(pourcentage_manioc_exporte), "%")
```

```python
# On calcule la quantité de manioc exporté dans le monde
quantite_manioc_exporte_monde = dispo_alimentaire['Exportations - Quantité'][dispo_alimentaire['Produit'] == 'Manioc'].sum()

#dispo_alimentaire.loc[(dispo_alimentaire['Produit'] == "Manioc")].sum()
print('Quantité de manioc exporté dans le monde : ', "{:.3f}".format(quantite_manioc_exporte_monde / 1000000), "millions de kilos", end="\n\n")
```

```python
# On calcule la proportion exportée de Thaïlande par rapport à l'exportation mondiale
pourcentage_manioc_thailande_monde = quantite_manioc_exporte_thailande * 100 / quantite_manioc_exporte_monde

print("{:.2f}".format(pourcentage_manioc_thailande_monde), "% du manioc exportés dans le monde provient de Thaïlande")
```

```python
# Pourcentage de personnes en sous-nutrition en Thaïlande qui auraient pu être nourries avec le manioc exporté
kcal_moyen_manioc = 1590
print("Le nombre de calories moyen pour un kilo de manioc : ", kcal_moyen_manioc, " kcal", end="\n\n")

# On calcule le nombre de kilos de manioc nécessaire pour un an pour une personne
kilos_manioc_an_personne = moyenne_kcal/kcal_moyen_manioc * 365
print("Quantité de manioc nécessaire pour une personne par an : ", "{:.2f}".format(kilos_manioc_an_personne), "kilos", end="\n\n")

# On calcule le nombre de personnes qui peuvent être nourries avec le manioc exporté
personnes_nourries_manioc_exportes = quantite_manioc_exporte_thailande / kilos_manioc_an_personne
print("Nombre de personnes pouvant être nourries avec le manioc exporté", personnes_nourries_manioc_exportes, end="\n\n")

# On calcule le pourcentage de personnes en sous-nutrition en Thaïlande qui auraient pu être nourries avec le manioc exporté
pourcentage_sous_nutrition_manioc = personnes_nourries_manioc_exportes / thailande_df['sous_nutrition'].iloc[0] * 100
print("Pourcentage de personnes en sous-nutrition en Thaïlande qui auraient pu être nourries avec le manioc exporté :", "{:.2f}".format(pourcentage_sous_nutrition_manioc), "%", end="\n\n")
```

```python
# Création d'une liste avec les 2 pays qui ont la plus grosse utilisation des aliments pour le bétail
liste_deux_pays_aliments_animaux = dispo_cereales_animaux['Zone'].head(2).tolist()

# Création d'un tableau avec les données du premier pays
cereales_premier_pays = cereales.loc[cereales['Zone'] == liste_deux_pays_aliments_animaux[0], :]
cereales_premier_pays = cereales_premier_pays.sort_values('Aliments pour animaux', ascending=False).reset_index()

produit_cereales_premier_pays = cereales_premier_pays['Produit'].iloc[0]
quantite_cereales_premier_pays = cereales_premier_pays['Aliments pour animaux'].iloc[0]

print("Le premier pays qui a la plus grosse utilisation des aliments pour le bétail est :", liste_deux_pays_aliments_animaux[0], "avec", quantite_cereales_premier_pays, "kilos de", produit_cereales_premier_pays, "utilisés", end="\n\n")

# Création d'un tableau avec les donénes du deuxième pays
cereales_second_pays = cereales.loc[cereales['Zone'] == liste_deux_pays_aliments_animaux[1], :]
cereales_second_pays = cereales_second_pays.sort_values('Aliments pour animaux', ascending=False).reset_index()

produit_cereales_second_pays = cereales_premier_pays['Produit'].iloc[1]
quantite_cereales_second_pays = cereales_premier_pays['Aliments pour animaux'].iloc[1]

print("Le deuxième pays qui a la plus grosse utilisation des aliments pour le bétail est :", liste_deux_pays_aliments_animaux[1], "avec", quantite_cereales_second_pays, "kilos de", produit_cereales_second_pays, "utilisés", end="\n\n")
```

```python
# Nombre de kcal pour le maïs
# Source : https://www.passeportsante.net/fr/Nutrition/EncyclopedieAliments/Fiche.aspx?doc=mais
kcal_mais_entier = 3460
kcal_mais_doux = 1060

# On calcule le nombre de kilos de mais nécessaire pour un an pour une personne
kilos_mais_entier_an_personne = moyenne_kcal/kcal_mais_entier * 365
print("Quantité de maïs entier nécessaire pour une personne par an : ", "{:.2f}".format(kilos_mais_entier_an_personne), "kilos", end="\n\n")
kilos_mais_doux_an_personne = moyenne_kcal/kcal_mais_doux * 365
print("Quantité de maïs doux nécessaire pour une personne par an : ", "{:.2f}".format(kilos_mais_doux_an_personne), "kilos", end="\n\n")
```

```python
# On calcule le nombre de personnes qui peuvent être nourries avec 10% du maïs servant pour le bétail
personnes_nourries_mais_entier_exportes = (quantite_cereales_premier_pays*10/100) / kilos_mais_entier_an_personne
print("Nombre de personnes pouvant être nourries avec 10% du maïs entier servant pour le bétail :", int(personnes_nourries_mais_entier_exportes), end="\n\n")
personnes_nourries_mais_doux_exportes = (quantite_cereales_premier_pays*10/100) / kilos_mais_doux_an_personne
print("Nombre de personnes pouvant être nourries avec 10% du maïs doux servant pour le bétail :", int(personnes_nourries_mais_doux_exportes), end="\n\n")
```
